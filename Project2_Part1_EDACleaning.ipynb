```python
#This function will be used for plotting countplots for comparison.
def subplot_countplot(dataframe,list_columns):
    '''plotting countplot function'''
    nrows = int(np.ceil(len(list_columns)/2))
    fig, ax = plt.subplots(nrows=nrows,ncols=2, figsize = (20,10*nrows))
    ax = ax.ravel()
    if len(list_columns) %2 !=0:
        ax[len(list_columns)].set_visible(False)
    for i, column in enumerate(list_columns):
        names = train.groupby(column).mean()['saleprice'].sort_values().index
        sns.countplot(dataframe[column], ax=ax[i])
        ax[i].set_title(f' "{column}"', fontsize =20)
        ax[i].set_ylabel('Count',fontsize=18)
        ax[i].set_xlabel(column,fontsize=18) 
        ax[i].set_xticklabels(names[::-1],rotation=40,fontsize=15) 
        ax[i].tick_params(axis='y',labelsize=15)
        
    plt.tight_layout()
```

```python
#Binazring all ordinal columns

#External Quality
train['exter qual'] = train['exter qual'].map({'Ex': 5,'Gd': 4,'TA': 3, 'Fa':2,'Po':1})
#External Condition
train['exter cond'] = train['exter cond'].map({'Ex': 5,'Gd': 4,'TA': 3, 'Fa':2,'Po':1})
#Basement Quality
train['bsmt qual'] = train['bsmt qual'].map({'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1,'NA':0})
#Basement Condition
train['bsmt cond'] = train['bsmt cond'].map({'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1,'NA':0})
#Basement Exposure
train['bsmt exposure'] = train['bsmt exposure'].map({'Gd': 4,'Av': 3,'Mn': 2,'No': 1,'NA':0})
#Basement Finish Type 1
train['bsmtfin type 1'] = train['bsmtfin type 1'].map({'GLQ': 6,'ALQ': 5,'BLQ': 4,'Rec': 3,'LwQ':2,'Unf':1,'NA':0})
#Basement Finish Type 2
train['bsmtfin type 2'] = train['bsmtfin type 2'].map({'GLQ': 6,'ALQ': 5,'BLQ': 4,'Rec': 3,'LwQ':2,'Unf':1,'NA':0})
#Heating QC
train['heating qc'] = train['heating qc'].map({'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1})
#Kitchen Quality
train['kitchen qual'] = train['kitchen qual'].map({'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1})
#Functional
train['functional'] = train['functional'].map({'Typ': 8,'Min1': 7,'Min2': 6,'Mod': 5,'Maj1': 4,'Maj2':3,'Sev':2,'Sal':1})
#Garage Finish
train['garage finish'] = train['garage finish'].map({'Fin': 3,'RFn': 2,'Unf': 1,'NA':0})
#Garage Quality
train['garage qual'] = train['garage qual'].map({'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1,'NA':0})
#Garage Condition
train['garage cond'] = train['garage cond'].map({'Ex': 5,'Gd': 4,'TA': 3,'Fa': 2,'Po': 1,'NA':0})
```
